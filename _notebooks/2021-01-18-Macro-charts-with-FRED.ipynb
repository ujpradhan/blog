```python
from fredapi import Fred 

#fred = Fred("use personal FRED API") 
```

```python
#hide_input
fred = Fred("415f48282451239a5a115a96ea67dfbb") #use personal FRED API
```

```python
#setting date parameters for all charts
begin = '2017-12-31'
end = '2021-01-15'
```

```python
#Get rates data series
fedfunds = fred.get_series('DFF',
                           observation_start = begin,
                           obervation_end = end) #daily

GT2 = fred.get_series('DGS2',
                      observation_start = begin,
                      obervation_end = end) #daily

GT10 = fred.get_series('DGS10',
                     observation_start = begin,
                     obervation_end = end) #daily

GT30 = fred.get_series('DGS30',
                     observation_start = begin,
                     obervation_end = end) #daily
```

```python
#equities
spx = fred.get_series('SP500',
                     observation_start = begin,
                     obervation_end = end) #daily

nasdaq = fred.get_series('NASDAQCOM',
                     observation_start = begin,
                     obervation_end = end) #daily
```

```python
#Commodities
brent = fred.get_series('DCOILBRENTEU',
                        observation_start = begin,
                        obervation_end = end) #daily

wti = fred.get_series('DCOILWTICO',
                      observation_start = begin,
                      obervation_end = end) #daily

hh = fred.get_series('DHHNGSP',
                      observation_start = begin,
                      obervation_end = end) #daily

copper = fred.get_series('PCOPPUSDM',
                         observation_start = begin,
                         obervation_end = end) #monthly
```

```python
#currencies

usd_eur = fred.get_series('DEXUSEU',
                     observation_start = begin,
                     obervation_end = end) #daily

usd_gbp = fred.get_series('DEXUSUK',
                     observation_start = begin,
                     obervation_end = end) #daily

yen_usd = fred.get_series('DEXJPUS',
                     observation_start = begin,
                     obervation_end = end) #daily

yuan_usd = fred.get_series('DEXCHUS',
                     observation_start = begin,
                     obervation_end = end) #daily
```

### Rates

```python
fig, axes = plt.subplots(2,2, figsize = (12, 8))

# rotate and align x-labels so they look better
fig.autofmt_xdate()

axes[0, 0].plot(fedfunds)
axes[0, 0].set_title('Effective Fed Funds Rate')
axes[0, 0].set_ylabel('%')


axes[0, 1].plot(GT2)
axes[0, 1].set_title('Generic 2yr Treasury Yield')
axes[0, 1].set_ylabel('%')

axes[1, 0].plot(GT10)
axes[1, 0].set_title('Generic 10yr Treasury Yield')
axes[1, 0].set_ylabel('%')

axes[1, 1].plot(GT30)
axes[1, 1].set_title('Generic 30yr Treasury Yield')
axes[1, 1].set_ylabel('%')
```

### Equities

```python
fig, axes = plt.subplots(1, 2, figsize = (12, 4))

fig.autofmt_xdate()

axes[0].plot(spx)
axes[0].set_title('S&P 500')

axes[1].plot(nasdaq)
axes[1].set_title('NASDAQ Composite')
```

### Commodities

```python
fig, axes = plt.subplots(2, 2, figsize = (12, 8))

fig.autofmt_xdate()

axes[0, 0].plot(brent)
axes[0, 0].set_title('Brent Crude Oil')
axes[0, 0].set_ylabel('$/bbl')

axes[0, 1].plot(wti)
axes[0, 1].set_title('WTI Crude Oil')
axes[0, 1].set_ylabel('$/bbl')

axes[1, 0].plot(hh)
axes[1, 0].set_title('Henry Hub Natural Gas')
axes[1, 0].set_ylabel('$/MMBtu')

axes[1, 1].plot(copper)
axes[1, 1].set_title('Copper')
axes[1, 1].set_ylabel('$/MT')
```

### Currencies

```python
fig, axes = plt.subplots(2, 2, figsize = (12, 8))

fig.autofmt_xdate()

axes[0, 0].plot(usd_eur)
axes[0, 0].set_title('USD/EUR Exchange Rate')

axes[0, 1].plot(usd_gbp)
axes[0, 1].set_title('USD/GBP Exchange Rate')

axes[1, 0].plot(yen_usd)
axes[1, 0].set_title('JPY/USD Exchange Rate')

axes[1, 1].plot(yuan_usd)
axes[1, 1].set_title('CNY/USD Exchange Rate')
```
